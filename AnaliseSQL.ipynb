## Número de Indenizações por tipo de Sinistro

Para isto irei utilizar a Tabela SINREG 

```python
sql = '''
select  "Tipo_Sinistro", SUM("Numero_Indenizacoes") as "Numero_Indenizacoes"
FROM "Seguro"."SINREG"
group by "Tipo_Sinistro"
order by SUM("Numero_Indenizacoes") DESC
'''
df = sqlio.read_sql_query(sql, con)
```

```python
df
```

Casco é o tipo de sinistro que possui mais indenizações, seguido de RCDM (Responsabilidade Civil Facultativa de Veículos – Danos Materiais)

## Número de Indenizações por Região

Para isto foi utilizada a Tabela SINREG

```python
sql = '''
select  "Descricao" as "Regiões", SUM("Numero_Indenizacoes") as "Numero_Indenizacoes"
FROM "Seguro"."SINREG"
group by "Descricao"
order by SUM("Numero_Indenizacoes") DESC
'''
df = sqlio.read_sql_query(sql, con)
```

```python
df
```

As regiões Metropolitana de São Paulo, Ribeirão Preto e demias Municípios de Campinas foram as que obtiveram maior número de indenizações.


## Roubo e Furto por Categoria Tarifária

Para isto iremos utilizar as Tabelas TARIFA e ARQ_CASCO

```python
sql = '''
select  "TARIFA"."Categoria", SUM("ARQ_CASCO"."Roubo/Furto") as "Roubo/Furto"
FROM "Seguro"."TARIFA", "Seguro"."ARQ_CASCO" 
where "TARIFA"."id_tarifa" = "ARQ_CASCO"."id_tarifa"
group by "Categoria"
order by  SUM("ARQ_CASCO"."Roubo/Furto") DESC
'''
df = sqlio.read_sql_query(sql, con)
```

```python
df
```

Vamos que a maioria dos roubos e furtos acontecem com veículos de passeio nacional, pick-up (nacional e importadas) e motocicletas

## Roubo e Furto por Região

Para isto iremos utilizar as Tabelas REGIAO e ARQ_CASCO

```python
sql = '''
select  "REGIAO"."Descricao" as "Regiões", SUM("ARQ_CASCO"."Roubo/Furto") as "Roubo/Furto"
FROM "Seguro"."REGIAO", "Seguro"."ARQ_CASCO" 
where "REGIAO"."id_regiao" = "ARQ_CASCO"."id_regiao"
group by "REGIAO"."Descricao"
order by  SUM("ARQ_CASCO"."Roubo/Furto") DESC
'''
df = sqlio.read_sql_query(sql, con)
```

```python
df
```

Temos que as regiões que mais acontecem Roubo e Furto são as regiões Metropolitanas de São Paulo e Rio de Janeiro.

## Roubo e Furto por Região e Categoria Tarifária

Para isto irei utilizar as Tabelas REGIAO, ARQ_CASCO e TARIFA

```python
sql = '''
select  "REGIAO"."Descricao" as "Regiões",  "TARIFA"."Categoria", SUM("ARQ_CASCO"."Roubo/Furto") as "Roubo/Furto"
FROM "Seguro"."REGIAO", "Seguro"."ARQ_CASCO", "Seguro"."TARIFA"
where "REGIAO"."id_regiao" = "ARQ_CASCO"."id_regiao" and "TARIFA"."id_tarifa" = "ARQ_CASCO"."id_tarifa"
group by "REGIAO"."Descricao", "TARIFA"."Categoria"
order by  SUM("ARQ_CASCO"."Roubo/Furto") DESC
'''
df = sqlio.read_sql_query(sql, con)
```

```python
df
```

A categoria Passeio nacional é a que sofre mais furtos na Região Metropolitana de São Paulo, seguida de Motocicleta.

## Roubo e Furto por Faixa Etária e Sexo

Para isto irei utilizar as Tabelas IDADE, SEXO e ARQ_CASCO

```python

sql = '''
select  "IDADE"."Descricao" as "Faixa_Etaria","SEXO"."Descricao" as "Sexo", SUM("ARQ_CASCO"."Roubo/Furto") as "Roubo/Furto"
FROM "Seguro"."IDADE", "Seguro"."ARQ_CASCO", "Seguro"."SEXO"
where "IDADE"."id_idade" = "ARQ_CASCO"."id_idade" AND "SEXO"."id_sexo" = "ARQ_CASCO"."id_sexo"
group by "IDADE"."Descricao", "SEXO"."Descricao"
order by SUM("ARQ_CASCO"."Roubo/Furto") DESC
'''
df = sqlio.read_sql_query(sql, con)
```

```python
df
```

Vemos que em todas as faixa etárias, Roubo e Furto acontecem mais com homens.

## Colisão Parcial e Perda Total por Faixa Etária e Sexo

Para isto irei utilizar as Tabelas IDADE, SEXO e ARQ_CASCO

```python
sql = '''
select  "IDADE"."Descricao" as "Faixa_Etaria","SEXO"."Descricao" as "Sexo", SUM("ARQ_CASCO"."Colisao_Parcial") as "Colisao_Parcial", SUM("ARQ_CASCO"."Perda_Total") as "Perda_Total"
FROM "Seguro"."IDADE", "Seguro"."ARQ_CASCO", "Seguro"."SEXO"
where "IDADE"."id_idade" = "ARQ_CASCO"."id_idade" AND "SEXO"."id_sexo" = "ARQ_CASCO"."id_sexo"
group by "IDADE"."Descricao", "SEXO"."Descricao"
order by SUM("ARQ_CASCO"."Colisao_Parcial") DESC, SUM("ARQ_CASCO"."Perda_Total") DESC
'''
df = sqlio.read_sql_query(sql, con)
```

```python
df
```

Aqui vemos que homens tendem a ter mais colisões parcicais e perdas totais que mulheres, em todas as faixas etárias.
